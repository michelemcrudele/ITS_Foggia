# Introduzione e generalit

Quando si fa unanalisi di dati, il modo in cui si mostrano i risultati  fondamentale:un buon grafico ti permette di capire subito se lanalisi ha senso e di comunicarla ad altri in modo efficace. Esempi famosi 
* la scoperta del [bosone di Higgs](https://twiki.cern.ch/twiki/pub/CMSPublic/Hig13002TWiki/HZZ4l_animated.gif)

In generale, un buon grafico deve trasmettere pi informazioni possibile in modo intuitivo e facile da leggere.



Noi oggi ci concentreremo su : 
   - **istogrammi** --> per capire come sono distribuiti i dati (unidimensionale)
   - **scatter plots** (grafici a dispersione)--> per studiare la relazione tra due variabili (bidimensionale)

------------------------------------------------------------------
### Istogrammi (Histograms)

Un istogramma serve per rappresentare la distribuzione di un insieme di dati.
Funziona cos:

- Si divide lintervallo dei dati in contenitori (chiamati bin).

- Si conta quanti dati finiscono in ogni contenitore.

- Si disegna un rettangolo per ogni bin, alto quanto il numero di dati contenuti.

Esempio: se stai misurando laltezza di un gruppo di persone, un istogramma pu dirti quanti stanno tra 150-160 cm, 160-170 cm, ecc. 


---------------------------------------------------------------

### Scatter plots

Gli scatter plot servono a mostrare la relazione tra due variabili.
Ogni punto rappresenta una coppia di valori (x, y).
Sono molto utili per capire se esiste una correlazione tra le due grandezze.

Esempio: vuoi sapere la relazione tra il numero di ore di studio e il voto di un esame 





### Box plot

Spesso si riassume lo scatter plot in un profilo:

- Si dividono i dati in bin lungo lasse x.

- Per ogni bin si calcola la media e la deviazione standard dei valori di y.

- Si rappresentano questi valori come punti con barre di errore.
Questo aiuta a capire landamento medio dei dati in modo pi pulito.

### Struttura di un grafico in Matplotlib

Un grafico in Matplotlib  composto da oggetti annidati:

- **Figure**   il contenitore principale, pu contenere uno o pi grafici.

- **Axes**  rappresentano i veri e propri grafici (quello che noi chiamiamo plot).

Nome un po fuorviante: Axes non significa assi, ma **larea in cui disegni il grafico**.

- Dentro un **Axes** troviamo altri elementi pi piccoli:

    - **Linee, punti, legende, etichette, tick** degli assi, ecc.

    - Ognuno di questi  un oggetto che si pu modificare (colore, spessore, testo, ecc.).

In pratica:

- **Figure** = la pagina

- **Axes** = il singolo disegno sulla pagina

- Dentro lAxes ci sono tutte le parti del grafico che puoi personalizzare

### Stateful vs Stateless Approaches

Quando facciamo grafici con Matplotlib, ci sono due modi principali per farlo: stateful (basato sullo stato) e stateless / object-oriented (basato sugli oggetti).

#### 1) Stateful (basato sullo stato)
Questo  il modo pi semplice.

Esempio tipico:

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Grafico semplice")
plt.show()
```

Cosa succede qui:
- plt.plot() lavora sempre sulla figura corrente.
- Se non esiste una figura, la crea automaticamente.
- Non dobbiamo preoccuparci di dire su quale figura/asse sto lavorando: Python lo fa da solo.

 In altre parole,  come una macchina con memoria: ricorda su quale grafico stai lavorando.


#### 2) Stateless / Object-Oriented (basato sugli oggetti)

Qui lavoriamo direttamente sugli oggetti che rappresentano il grafico (Figure e Axes).


```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()  # Creo una figura e un asse
ax.plot([1, 2, 3], [4, 5, 6])
ax.set_title("Grafico con oggetti")
plt.show()
```

Cosa succede qui:
- fig  la figura.
- ax  lasse (dove c il grafico vero e proprio).
- Possiamo modificare tutto direttamente su ax, senza affidarsi alla memoria di pyplot.

 Questo  utile quando facciamo grafici complessi o tanti grafici insieme.

```python
fig, axs = plt.subplots(1, 2, figsize=(10,4))

# Primo grafico
axs[0].plot(x, y, color='red', marker='o')
axs[0].set_title("Grafico Lineare")

# Secondo grafico
axs[1].hist(data, bins=15, color='purple', edgecolor='black')
axs[1].set_title("Istogramma")

plt.show()
```

### A few simple examples

```python
## Creo un intervallo di anni
rng = np.arange(50) # numeri da 0 a 49
yrs = 1950 + rng # trasformo in anni dal 1950 al 1999

## Creo dati casuali per tre regioni
rnd = np.random.randint(0, 10, size=(3, rng.size)) 

# Creo la figura e l'asse
fig, ax = plt.subplots(figsize=(10, 6))

# NB: Tutti i metodi sono chiamati direttamente su ax (approccio object-oriented)
ax.stackplot(yrs, rng + rnd, labels=['Eastasia', 'Eurasia', 'Oceania']) # disegna le aree sovrapposte
ax.set_title('Combined debt growth over time') # aggiunge titolo
ax.legend(loc='upper left') # aggiunge legenda
ax.set_ylabel('Total debt') # aggiunge etichetta asse y
ax.set_xlim(left=yrs[0], right=yrs[-1]) # limiti asse x

plt.show()
```

```python
# Creo i dati
x = np.random.randint(1, 11, size=50) # 50 numeri casuali tra 1 e 10
y = x + np.random.randint(1, 5, size=x.size)  # y  x pi un piccolo rumore casuale

# Creo una figura con due grafici affiancati
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4)) # grafici messi su 1 riga e 2 colonne

# Grafico scatter sul primo asse
ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='b')  # punti rossi con bordo blu
ax1.set_title('Scatter: x vs y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# Istogramma sul secondo asse
data = np.column_stack((x, y))  # combina x e y in un array
ax2.hist(x, bins=np.arange(data.min(), data.max()), label='x', alpha=0.5)
ax2.hist(y, bins=np.arange(data.min(), data.max()), label='y', alpha=0.5)
ax2.legend(loc=(0.65, 0.8))  # legenda posizionata dentro il grafico
ax2.set_title('Frequencies of x and y')
ax2.yaxis.tick_right()       # mette i numeri dell'asse y a destra

# Mostro il grafico
plt.show()
```

Cosa succede in pratica
- Scatter plot (ax1)
    - Ogni punto rappresenta una coppia (x, y).
    - Colore e bordo servono a distinguere i punti visivamente.

- Istogramma (ax2)
    - Mostra quante volte ogni valore di x e y appare nei dati.
    - Usando alpha=0.5, le barre diventano semi-trasparenti e si vedono anche le sovrapposizioni.

- Figura con due assi
    - fig, (ax1, ax2) crea due grafici affiancati in una sola figura.
    - Lapproccio object-oriented permette di modificare ogni grafico indipendentemente.

### A more advanced example

```python
## Dati del mercato immobiliare in California

import numpy as np
from io import BytesIO
import tarfile
from urllib.request import urlopen

# Scarico larchivio dal web
url = 'http://www.dcc.fc.up.pt/~ltorgo/Regression/cal_housing.tgz'
b = BytesIO(urlopen(url).read()) # leggo larchivio in memoria

# Percorso del file dati dentro larchivio
fpath = 'CaliforniaHousing/cal_housing.data'

# Apro larchivio e leggo i dati
with tarfile.open(mode='r', fileobj=b) as archive:
    housing = np.loadtxt(archive.extractfile(fpath), delimiter=',')

# Controllo quante colonne ci sono
len(housing[0])

# Estraggo le colonne di interesse
z = housing[:, -1]      # tutte le righe, ultima colonna: valore medio delle case 
pop = housing[:, 4]     # colonna 4: popolazione
age = housing[:, 7]     # colonna 7: et media delle case

# Creo la figura e il layout dei grafici
fig = plt.figure(figsize=(12, 8))
gridsize = (3, 2) #3 righe, 2 colonne

# Scatter plot principale 
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=3, rowspan=2) #(0,0)-> parte da 1riga1col ; colspan=3 ->occupa 3colonne; rowspan=2 -> occupa 2righe
scatter = ax1.scatter(x=age, y=pop, c=z, cmap='RdYlGn')
plt.colorbar(scatter, ax=ax1, format='$%d')
ax1.set_title('Home value as a function of home age & area population', fontsize=14)
ax1.set_yscale('log')
ax1.set_xlabel('Home age')
ax1.set_ylabel('Population')


# Istogramma dell'et media delle case
ax2 = plt.subplot2grid(gridsize, (2, 0)) #parte da 3 rig, 1 col
ax2.hist(age, bins='auto') # matplotlib decide automaticamnete numero intervalli
ax2.set_title('Distribution of home age')
ax2.set_xlabel('Age')
ax2.set_ylabel('Count')

# Istogramma della popolazione (asse y log)
ax3 = plt.subplot2grid(gridsize, (2, 1)) #2col
ax3.hist(pop, bins='auto', log=True)
ax3.set_title('Distribution of population')
ax3.set_xlabel('Population')
ax3.set_ylabel('Count (log scale)')

plt.tight_layout()
plt.show()
```
