```python
# Esercizio 1 – Istogramma base
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
heights = np.random.randint(150, 200, size=50)

plt.hist(heights, bins=5, color='skyblue', edgecolor='black')
plt.title("Distribuzione delle altezze")
plt.xlabel("Altezza (cm)")
plt.ylabel("Numero di studenti")
plt.show()
```

```python
# Esercizio 2 – Modifica bin e colori
plt.hist(heights, bins=10, color='orange', edgecolor='black')
plt.title("Distribuzione delle altezze (10 bin)")
plt.xlabel("Altezza (cm)")
plt.ylabel("Numero di studenti")
plt.show()
```

```python
# Esercizio 3 – Scatter plot semplice
x = np.random.randint(1, 11, size=20)
y = x + np.random.randint(1, 5, size=x.size)

plt.scatter(x, y, color='red', edgecolor='blue')
plt.title("Scatter plot x vs y")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Esercizio 4 – Scatter plot con color map
z = np.random.randint(0, 101, size=20)  # terza variabile per colore

plt.scatter(x, y, c=z, cmap='viridis', s=100)
plt.colorbar(label='Valore colorato')
plt.title("Scatter plot con colormap")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Esercizio 5 – Box plot
data1 = np.random.randint(1, 20, size=15)
data2 = np.random.randint(5, 25, size=15)

plt.boxplot([data1, data2], labels=['Data1', 'Data2'], patch_artist=True)
plt.title("Box plot di due dataset")
plt.show()
```

```python
# Esercizio 6 – Più grafici nella stessa figura
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
ax1.plot(x, y1, color='red')
ax1.set_title("Seno")
ax2.plot(x, y2, color='blue')
ax2.set_title("Coseno")
plt.show()
```

```python
# Esercizio 7 – Istogramma log scale
data = np.random.randint(10, 10000, size=50)
plt.hist(data, bins='auto', color='green', edgecolor='black', log=True)
plt.title("Istogramma con asse Y logaritmico")
plt.xlabel("Valore")
plt.ylabel("Frequenza (log scale)")
plt.show()
```

```python
# Esercizio 8 – Scatter con dimensione e colore variabili
x = np.random.rand(30)
y = np.random.rand(30)
sizes = np.random.randint(20, 200, size=30)
colors = np.random.rand(30)

plt.scatter(x, y, s=sizes, c=colors, cmap='plasma', alpha=0.7)
plt.colorbar(label='Valore colore')
plt.title("Scatter plot con dimensione e colore variabili")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Esercizio 10 – Mini-progetto completo
age = np.random.randint(18, 30, size=30)        # età studenti
books = np.random.randint(0, 50, size=30)      # numero libri letti
grades = np.random.randint(60, 100, size=30)   # voto medio

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

# Istogramma età
axes[0].hist(age, bins='auto', color='lightblue', edgecolor='black')
axes[0].set_title("Distribuzione età")
axes[0].set_xlabel("Età")
axes[0].set_ylabel("Conteggio")

# Scatter libri vs voto
scatter = axes[1].scatter(books, grades, c=age, cmap='viridis', s=100)
axes[1].set_title("Libri letti vs Voto medio")
axes[1].set_xlabel("Libri letti")
axes[1].set_ylabel("Voto medio")
plt.colorbar(scatter, ax=axes[1], label='Età')

# Box plot voti
axes[2].boxplot(grades, patch_artist=True)
axes[2].set_title("Box plot dei voti")

plt.tight_layout()
plt.show()
```

- Con un’unica figura, mostra:
  1. **Istogramma** di `total_bill` con `kde=True`
  2. **Scatter plot** di `total_bill` vs `tip` con `hue="day"` e `size="size"`
  3. **Heatmap** delle correlazioni tra tutte le variabili numeriche
- Usa `plt.figure(figsize=(15,5))` per avere una figura grande


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carico il dataset tips
tips = sns.load_dataset("tips")
tips.head()
```

```python
# 1

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker")
plt.title("Mancia in funzione del conto totale (colorato per fumatore)")
plt.show()
```

```python
#2

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker", style="time")
plt.title("Mancia vs Conto - Forma in base a pranzo/cena")
plt.xlabel("Totale conto")
plt.ylabel("Mancia")
plt.show()
```

```python
# 3

sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips)
plt.suptitle("Mancia vs Conto - Dimensione in base al numero di persone", y=1.02)
plt.show()
```

```python
# 4

# Istogramma semplice
sns.histplot(data=tips, x="total_bill")
plt.title("Distribuzione del conto totale")
plt.show()

# Istogramma con KDE
sns.histplot(data=tips, x="total_bill", kde=True)
plt.title("Distribuzione del conto totale + KDE")
plt.show()

# Istogramma con KDE e element="step"
sns.histplot(data=tips, x="total_bill", kde=True, element="step")
plt.title("Distribuzione del conto totale + KDE (step)")
plt.show()
```

```python
# 5

# Scatter
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter")
plt.suptitle("Jointplot Scatter: Mancia vs Conto", y=1.02)
plt.show()

# KDE
sns.jointplot(data=tips, x="total_bill", y="tip", kind="kde", fill=True, cmap="Blues")
plt.suptitle("Jointplot KDE: Zone di densità", y=1.02)
plt.show()
```

```python
#6

sns.pairplot(tips, hue="sex")
plt.suptitle("Matrice di Scatterplot (colorata per sesso)", y=1.02)
plt.show()
```

```python
#7

corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Matrice di correlazione")
plt.show()
```

```python
# 8

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Istogramma con KDE
sns.histplot(data=tips, x="total_bill", kde=True, ax=axes[0])
axes[0].set_title("Distribuzione conto totale")

# Scatter plot con hue e size
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", size="size", ax=axes[1])
axes[1].set_title("Mancia vs Conto (giorno & numero persone)")

# Heatmap correlazioni
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axes[2])
axes[2].set_title("Correlazioni variabili numeriche")

plt.tight_layout()
plt.show()
```
