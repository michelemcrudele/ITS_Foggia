# Seaborn

Seaborn è una libreria basata su Matplotlib.

Serve a creare grafici più belli e facili da leggere.

Ha funzioni avanzate per analisi statistiche, come distribuzioni, correlazioni e scatter plot con regressione.

In pratica, Seaborn fa tutto quello che Matplotlib fa, ma con meno codice e grafici più accattivanti.

```python
import seaborn as sns
#sns.set(style="darkgrid")
```

### Scatter Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

# dataset già incluso nella libreria
tips = sns.load_dataset("tips")
tips.head()

# Il dataset tips contiene dati su mance in un ristorante, con colonne come:
# total_bill → totale del conto
# tip → mance
# sex → sesso del cliente
# smoker → fumatore/non fumatore
# day → giorno della settimana
# time → pranzo/cena
# size → numero di persone al tavolo
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker")
# equivalente: sns.relplot

# hue--> Cambia il colore dei punti in base a una colonna del dataset.

plt.show()
```

Qui Seaborn crea automaticamente colori, legenda e stili senza doverli impostare manualmente.

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips);

# style --> Cambia la forma del marker (cerchio, triangolo, quadrato) in base a una colonna.
```

Not only categorical but also quantitative "3rd dimension" is of course possible

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);

# hue --> quando vuoi distinguere categorie con colori.
```

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips);

# size --> quando vuoi rappresentare una variabile numerica in modo proporzionale alla grandezza del punto
# sizes=(15, 200) → dice: il punto più piccolo deve avere dimensione 15; quello più grande deve avere dimensione 200; tutti i valori intermedi vengono scalati proporzionalmente tra 15 e 200.
```

### Histograms

Un istogramma conta quante volte un certo valore (o intervallo di valori) compare nei dati.
In Seaborn hai diverse funzioni per disegnarlo, la più usata è:

```python
sns.histplot(data=tips, x="total_bill")
```

Ma Seaborn può fare molto di più rispetto a plt.hist() di Matplotlib:
- Grafica più bella già di default.
- Stime statistiche aggiuntive come la KDE 

La KDE (Kernel Density Estimate) è una curva liscia che rappresenta la distribuzione continua dei dati. In pratica, invece di vedere solo le barre dell’istogramma, hai una curva che ti dice dove i dati sono più concentrati.

```python
sns.histplot(data=tips, x="total_bill", kde=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill")
plt.show()
```

```python
# with unbinned distribution too

sns.histplot(data=tips, x="total_bill", kde=True, element="step")
plt.show()

# "step" si può cambiare in:
# "bars" → barre normali (default)
# "step" → disegna un contorno a gradini invece di barre piene
# "poly" → disegna un poligono pieno collegando i vertici
```

### Scatter plots and projections

`sns.jointplot()`:
- Mostra un grafico a dispersione di due variabili (es. x e y).
- Disegna anche le distribuzioni marginali (cioè gli istogrammi di x e y separati) sui lati.

In pratica, hai tutte le informazioni in un colpo d’occhio:
- Al centro: relazione tra x e y.
- Sopra: distribuzione di x.
- A destra: distribuzione di y.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset di esempio
tips = sns.load_dataset("tips")

# Jointplot di total_bill (conto totale) e tip (mancia)
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter")

plt.show()

# kind -> grafico centrale
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex")

# hex --> punti esagonali
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg")

# reg --> punti + linea regressione che mostra la tendenza
```

```python

# Carico il dataset Iris
iris = sns.load_dataset("iris")

g = sns.jointplot(x="sepal_width", y="petal_length", data=iris, kind="kde", space=0, cmap="Greens", fill=True)
plt.show()
```

### Ricerca delle correlazioni

Seaborn ti consente di cercare correlazioni controllando sistematicamente a coppie le feature del tuo dataset 


```python
sns.pairplot(iris);

# Crea un matrice di scatter plot per tutte le combinazioni di colonne numeriche.
# Sulla diagonale, invece dello scatter plot, mette istogrammi o KDE (distribuzione) di ogni variabile.
# Invece di creare molti grafici a mano, pairplot fa tutti i confronti in un’unica figura.
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```
